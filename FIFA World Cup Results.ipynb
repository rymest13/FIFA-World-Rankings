```python
# using df.loc to search for FIFA World Cup Tournament teams only
df_fwc = df.loc[df['tournament'] == 'FIFA World Cup']
df_fwc
```

```python
# point difference column
# if point difference is negative - away team won
# if point difference is positive - home team won
# if point difference is 0 - tie game

h = df_fwc['home_team'].value_counts()
a = df_fwc['away_team'].value_counts()

print('Home team games played: \n'+str(h))
print('Away team games played: \n'+str(a))
```
