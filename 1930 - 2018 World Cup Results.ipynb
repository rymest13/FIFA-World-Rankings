```python
first = df['CHAMPION'].value_counts()
second = df['RUNNER_UP'].value_counts()
third = df['THIRD_PLACE'].value_counts()

print('Teams first place finishes: \n'+str(first))
print('Teams second place finishes: \n'+str(second))
print('Teams third place finishes: \n'+str(third))
```

```python
plt.hist(df['CHAMPION'], bins=9)
plt.xticks(rotation=45)
plt.title('World Cup Results')
plt.xlabel('Country')
plt.ylabel('Champion')
plt.show()

plt.hist(df['RUNNER_UP'], bins=11)
plt.xticks(rotation='vertical')
plt.title('World Cup Results')
plt.xlabel('Country')
plt.ylabel('Runner Up')
plt.show()

plt.hist(df['THIRD_PLACE'], bins=15)
plt.xticks(rotation='vertical')
plt.title('World Cup Results')
plt.xlabel('Country')
plt.ylabel('Third Place')
plt.show()
```

```python
groups = df.groupby('CHAMPION')
for name, group in groups:
    plt.plot(group['GOALS_SCORED'], group['AVG_GOALS_PER_GAME'], marker='o', linestyle="", label=name)
    plt.title('World Cup Champion by Total Goals/Avg Goals per Game')
    plt.legend(bbox_to_anchor = (1,1))
```
