```python
# Seeing if my data will fit linear regression

x = df['rank']
y = df['total_points']

slope, intercept, r, p, std_err = stats.linregress(x, y)

print(r)
```

```python
# I like the fit from this data, will continue using the 2015-2018 data

x = df['rank']
y = df['total_points']

slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
    return slope * x + intercept

mymodel = list(map(myfunc, x))

plt.scatter(x, y)
plt.plot(x, mymodel)
plt.show()
```
