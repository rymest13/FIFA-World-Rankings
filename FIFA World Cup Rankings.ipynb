```python
# using df.loc to search for the number 1 team by rank
df_number1 = df.loc[df['rank'] == 1]
df_number1
```

```python
# using df_number1 which has all the countries who have been number 1
# using value counts to see how many times they've been number 1

df_number1['country_full'].value_counts()
```

```python
plt.hist(df_number1['country_full'], bins=8)
plt.xticks(rotation=45)
plt.title('How many times a country has been number 1')
plt.xlabel('Country')
plt.ylabel('Number 1')
plt.show()
```

```python
plt.bar(df_number1['country_abrv'], df_number1['total_points'])
plt.title('Number 1 countries total points')
plt.xlabel('Country')
plt.ylabel('Total Points')
plt.show()
```
